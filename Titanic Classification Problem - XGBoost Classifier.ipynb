```python
# GETTING THE DATASET

import pandas as pd
import numpy as np

df = pd.read_csv('/home/angel/Downloads/titanic.csv')
```

```python
df
```

```python
df_filtred = df.drop(['PassengerId', 'Survived', 'Name', 'Ticket'], axis=1)
```

```python
df_filtred
```

```python
# At this point we start to preprocess the data

X = df_filtred
y = df['Survived'].to_numpy()
```

```python
X['Sex'] = X['Sex'].map({'male': 1, 'female': 0})
```

```python
X
```

```python
print(X['Embarked'].unique())
print(X['Cabin'].unique())
```

```python
len(X['Cabin'].unique())
```

```python
# Using an encoder

from sklearn.preprocessing import TargetEncoder

EmbEnc = TargetEncoder()
CabinEnc = TargetEncoder()

X_Cabin = pd.DataFrame(X['Cabin'])
X_Emb = pd.DataFrame(X['Embarked'])

CabinEnc.fit(X_Cabin, y)
EmbEnc.fit(X_Emb, y)

X_Cabin = CabinEnc.transform(X_Cabin)
X_Emb = EmbEnc.transform(X_Emb)
```

```python
X['Cabin'] = X_Cabin
X['Embarked'] = X_Emb
```

```python
X
```

```python
X.loc[X['Cabin'] == 1, 'Cabin'] = None
```
